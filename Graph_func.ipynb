```python
from matplotlib.cm import ScalarMappable

def graf_outliers(dataf, target, district, repair, square, years, dist, rep, year, figsize = (15, 10), s = 7, alpha = 0.6, year_check = 0):
    
    """
    dataf - DataFrame, из которого используются данные
    target - цена за 1 кв.м
    district - столбец с районами города
    repair - столбец с видами ремонта
    square - столбец с площадями квартир
    years - столбец с годами постройки домов
    dist - сортировка по району города, в рамках которого будет построен график
    rep - сортировка по ремонту, в рамках которого будет построен график
    year - сортировка по году постройки, в рамках которого будет построен график
    figsize - размер графика, задан по умолчанию (15, 10)
    s - размер точки, задан по умолчанию в размере 7
    alpha - прозрачность точки, задана по умолчанию в размере 0.6
    year_check - необходимость сортировки по году постройки, принимает значения 0 или 1, по умолчанию нет = 0
    """
    
    if year_check == 1:
        df_graf = dataf[dataf[district] == dist][(dataf[dataf[district] == dist][repair] == rep) & (dataf[dataf[district] == dist][years] == year)]
    elif year_check == 0:
        df_graf = dataf[dataf[district] == dist][(dataf[dataf[district] == dist][repair] == rep)]
    else:
        return print('year_check принимает значения только 0 или 1')
    
    if len(df_graf) < 2:
        return print('Количество точек (менее 2-х) не позволяет построить график!')
    
    n_colors_s = len(df_graf[df_graf[repair] == rep][square].value_counts(sort = True, ascending = False))
    n_colors_y = len(df_graf[df_graf[repair] == rep][years].value_counts(sort = True, ascending = False))


    mini_s = df_graf[df_graf[repair] == rep][square].min()
    maxi_s = df_graf[df_graf[repair] == rep][square].max()
    mini_y = df_graf[df_graf[repair] == rep][years].min()
    maxi_y = df_graf[df_graf[repair] == rep][years].max()

    if maxi_y == mini_y:
        mini_y = mini_y
        maxi_y = maxi_y + 1
        print(f'У всех квартир одинаковый год {int(mini_y)}!')

    if maxi_s == mini_s:
        maxi_s = mini_s
        maxi_s = mini_s + 1
        print(f'У всех квартир одинаковая площадь {mini_s}!')

    norm_s = plt.Normalize(
                           mini_s, 
                           maxi_s
                           )

    norm_y = plt.Normalize(
                           mini_y, 
                           maxi_y
                           )

    cmap = plt.get_cmap("Spectral") #Spectral, seismic

    #cpal_s = sns.color_palette('Spectral', n_colors = n_colors_s, desat = 1)
    #cpal_y = sns.color_palette('Spectral', n_colors = n_colors_y, desat = 1)

    fig, ax = plt.subplots(1, 2, figsize = figsize) #1 строка и 2 столбика, figsize = ширина и высота, sharex=True,

    #создаем boxplot
    sns.boxplot(
                x = district, 
                y = target, 
                data = df_graf[df_graf[repair] == rep],
                color = 'grey',
                ax = ax[0]
                ) 

    #создаем strip plot для площади
    sns.stripplot(
                  x = district, 
                  y = target,
                  hue = square,
                  #hue_norm = (mini, maxi), #делает нормализацию от мин до макс, как в norm выше
                  #orient = "h",
                  #col = "day",       #тут можно указать районы и не сортировать их
                  color = [cmap(norm_s(df_graf[df_graf[repair] == rep][square]))], #'black',
                  alpha = alpha, 
                  data = df_graf[df_graf[repair] == rep],
                  #jitter = (0, 0.65, 0.05), #True/False
                  #linewidth = 2,
                  #marker = '*',
                  palette = 'Spectral', #cpal_s, #Spectral, seismic
                  edgecolor = 'none',
                  s = s,
                  ax = ax[0]
                  )
    #sns.despine() #удаляет полосы с графика

    #добавляем цветовую шкалу
    sm_s = ScalarMappable(norm = norm_s, cmap = cmap)
    sm_s.set_array([])

    fig.colorbar(
                 sm_s, 
                 ax = ax[0], 
                 orientation = 'vertical', #horizontal
                 label = 'Площадь'
                 )

    #cbar.ax.set_title("scale")

    #создаем boxplot
    sns.boxplot(
                x = district, 
                y = target, 
                data = df_graf[df_graf[repair] == rep],
                color = 'grey',
                ax = ax[1]
                ) 

    #создаем strip plot для ГП
    sns.stripplot(
                  x = district, 
                  y = target,
                  hue = years,
                  #hue_norm = (mini, maxi), #делает нормализацию от мин до макс, как в norm выше
                  #orient = "h",
                  #col = "day",       #тут можно указать районы и не сортировать их
                  color = [cmap(norm_s(df_graf[df_graf[repair] == rep][years]))], #'black',
                  alpha = alpha, 
                  data = df_graf[df_graf[repair] == rep],
                  #jitter = (0, 0.65, 0.05), #True/False
                  #linewidth = 2,
                  #marker = '*',
                  palette = 'Spectral', #cpal_s, #Spectral, seismic
                  edgecolor = 'none',
                  s = s,
                  ax = ax[1]
                  )
    #sns.despine() #удаляет полосы с графика

    sm_y = ScalarMappable(norm = norm_y, cmap = cmap)
    sm_y.set_array([])

    fig.colorbar(
                 sm_y, 
                 ax = ax[1], 
                 orientation = 'vertical', #horizontal
                 label = 'Год постройки'
                 )

    ax[0].get_legend().remove() #отключаем легенду у площади
    ax[1].get_legend().remove() #отключаем легенду у ГП

    #plt.xticks(np.arange(0,151,25))
    ax[0].set_yticks(np.arange(df_graf[df_graf[repair] == rep][target].min() - 5000, 
                               df_graf[df_graf[repair] == rep][target].max() + 5000, 20000))
    ax[1].set_yticks(np.arange(df_graf[df_graf[repair] == rep][target].min() - 5000, 
                               df_graf[df_graf[repair] == rep][target].max() + 5000, 20000))
    
    ax[1].set_ylabel(None) #отключаем название оси у второго boxplot

    if year_check == 1:
        fig.suptitle(f'Распределение цен в районе {dist}, вид ремонта - {rep}, год постройки - {year}')
    else:
        fig.suptitle(f'Распределение цен в районе {dist}, вид ремонта - {rep}, все годы постройки')
    
    #plt.title("Распределение цен (западный обход, предчистовая)", size = 15)
    #plt.xlabel("Район", size = 14)
    #plt.ylabel("Цена(кв.м)", size = 14)
    return plt.show()
```

```python
def table_count(dataf, district, dist, repair, years, left, sort = 0):
    """
    По району можно получить количество квартир в зависимости от отделки и года постройки
    dataf - DataFrame, из которого используются данные
    district - столбец с районами города
    repair - столбец с видами ремонта
    square - столбец с площадями квартир
    years - столбец с годами постройки домов
    dist - сортировка по району города, в рамках которого будет построен график
    left - нижняя граница подсветки, диапазон вида [left, max]
    sort - сортировка итоговой таблицы по значению left
    """
    
    df_graf = dataf[dataf[district] == dist] #сортировка по району
    
    df_graf[years] = df_graf[years].astype(int)
    df_graf['ед'] = 1
    
    tb = df_graf.pivot_table(
                             values = 'ед',
                             index = years,
                             columns = repair,
                             aggfunc = 'count'
                            ).fillna(0)

    if sort == 1:
        tb['max'] = tb.max(axis = 1)
        tb = tb[tb['max'] >= left]
        tb = tb.drop('max', axis = 1)
    
    tb = tb.astype(int).style.highlight_between(left = left, right = np.inf, color = 'lightgreen')
    #tb = tb.append(tb[list(tb.columns)].sum(), ignore_index = True).fillna('')
    #print(df_graf[repair].value_counts())
    vc = df_graf[repair].value_counts()
    
    return vc, tb
```
