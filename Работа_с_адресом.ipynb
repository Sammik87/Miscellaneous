```python
street_types = [
    "улица", "ул", "ул.", "переулок", "пер", "пер.", "проспект", "пр", "пр.", "пр-т", "бульвар", "б-р", "площадь", "пл", "пл.",
    "тупик", "наб", "наб.", "набережная", "шоссе", "ш", "проезд", "жк", "мкр", "мкр.", "квартал", "поселок", "пос", "пос.",
    "район", "р-н", "аллея", "тракт", "просек"
]

street_pattern = (
    r"(?:ул\.?|проспект|просп\.?|переулок|проезд|шоссе|тупик|бульвар|наб\.?|линия)\s"  # Тип улицы
    r"(?:имени\s)?"
    r"[\dА-Яа-яЁё-]+(?:\s[\dА-Яа-яЁё-]+)*"  # Название улицы с возможными числами (1-я, 65-й)
)

def extract_address_parts(address):
    address = address.replace("\n", " ")
    address = address.lower().replace(",", " ").replace("  ", " ")  # Приводим к нижнему регистру, убираем запятые
    address_parts = address.split()  # Разбиваем на слова

    #print(address)

    street_name = None
    street_type = None
    house_number = None

    for i, word in enumerate(address_parts):
        if word in street_types:  # Нашли тип улицы
            street_type = word
            if i > 0:
                street_name = " ".join(address_parts[:i])  # Все слова до типа улицы — это название
            break

    if not street_name:  # Если не нашли тип улицы, пробуем забрать просто название
        for i in range(len(address_parts) - 1):
            if re.match(r'\d+', address_parts[i + 1]):  # Если следующее слово — число, значит это номер дома
                street_name = " ".join(address_parts[:i + 1])
                break

    # Ищем номер дома с учетом корпуса и дробей
    pattern = r'\b\d+[А-Яа-я]?(?:/\d+)?(?:к\d+)?(?:лит[А-Яа-я0-9])?(?:с\d+)?\b' #r'\d+[А-Яа-я]?(?:/\d+)?(?:к\d+)?(?:лит[А-Яа-я0-9])?(?:с\d+)?'
    match = re.search(pattern, address)
    if match:
        house_number = match.group(0)

    return street_name, street_type, house_number
```
